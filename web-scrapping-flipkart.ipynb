```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import time

final_df = pd.DataFrame()

# Iterate through multiple pages of a website to scrape smartphone data
for page in range(1, 41):

    # Define the URL and headers for making the request
    url = f'https://www.flipkart.com/search?q=smartphones&page={page}'
    header = {'User_Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

    # Send a GET request to the URL with the provided headers
    response = requests.get(url, headers=header)

    # Get the HTML content from the response
    result = response.content

    # Create a BeautifulSoup object to parse the HTML content
    soup = bs(result)

    # Find all the div elements with the class '_2kHMtA' which contain smartphone data
    all_div = soup.find_all('div', attrs={'class': '_2kHMtA'})

    # Initialize lists to store the scraped data
    original_price = []
    discounted_price = []
    image_url = []
    description = []
    total_ratings = []
    total_reviews = []
    avg_rating = []

    # Extract the required information from each div element
    for i in all_div:
        
        try:
            description.append(i.find('img', attrs={'class': '_396cs4'}).get('alt'))
        except:
            description.append(np.nan)
            
        try:
            image_url.append(i.find('img', attrs={'class': '_396cs4'}).get('src'))
        except:
            image_url.append(np.nan)
            
        try:
            total_ratings.append(i.find('span', attrs={"class": "_2_R_DZ"}).text.split()[0].replace(',', ''))
        except:
            total_ratings.append(np.nan)
            
        try:
            total_reviews.append(i.find('span', attrs={"class": "_2_R_DZ"}).text.split()[3].replace(',', ''))
        except:
            total_reviews.append(np.nan)
            
        try:
            avg_rating.append(i.find('div', attrs={'class': '_3LWZlK'}).text)
        except:
            avg_rating.append(np.nan)
            
        try:
            discounted_price.append(i.find('div', attrs={'class': '_30jeq3 _1_WHN1'}).text[1:].replace(',', ''))
        except:
            discounted_price.append(np.nan)
            
        try:
            original_price.append(i.find('div', attrs={'class': '_3I9_wc _27UcVY'}).text[1:].replace(',', ''))
        except:
            original_price.append(np.nan)

            
    # Create a DataFrame from the scraped data
    df = pd.DataFrame({
        'original_price': original_price,
        'discounted_price': discounted_price,
        'image_url': image_url,
        'description': description,
        'total_ratings': total_ratings,
        'total_reviews': total_reviews,
        'avg_rating': avg_rating
    })

    # Concatenate the current DataFrame with the final DataFrame
    final_df = pd.concat([final_df, df], ignore_index=True)

    # Pause execution for 3 seconds to avoid overwhelming the server
    time.sleep(3)

# Perform further operations on the final DataFrame if needed
```
